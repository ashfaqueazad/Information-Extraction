# Task 1 (10 Marks)
Write a function that takes each document and performs:
1) sentence segmentation 2) tokenization 3) part-of-speech tagging

Please keep in mind that the expected output is a list within a list as shown below.


```python
def ie_preprocess(document):
    #code goes here
    pos_sentences=[]#declare an empty list
    if(isinstance(document,list)):#check if it is a list
        for i in range(len(document)):
            sentences = nltk.sent_tokenize(document[i])#sentence tokenize
            sentences = [nltk.word_tokenize(sent) for sent in sentences] #word tokenize
            sentences = [nltk.pos_tag(sent) for sent in sentences] #part-of-speech tagging
            pos_sentences.append(sentences)#append it to the list
    else:# if it is not a list
        sentences = nltk.sent_tokenize(document)
        sentences = [nltk.word_tokenize(sent) for sent in sentences]
        pos_sentences = [nltk.pos_tag(sent) for sent in sentences]
    #returns a list of tokens with pos tags    
    return pos_sentences
```

# Task 2 (20 Marks)
Write a function that will take the list of tokens with POS tags for each sentence and returns the named entities (NE). 

Hint: Use binary=True while calling NE chunk function

```python
def named_entity_finding(pos_sent):
    #Code goes here
    tree=nltk.ne_chunk(pos_sent[0],binary=True)#Named entities will be tagged just as 'NE'
    named_entities=[]#empty list
    for subtree in tree.subtrees():
        if subtree.label()=='NE':#if the label is a named entity
            entity=""#empty string
            for leaf in subtree.leaves():
                entity =entity + leaf[0] + " "
                named_entities.append(entity.strip())
    #returns a list of named entities
    return named_entities

pos_sents=ie_preprocess(list_of_doc)
named_entity_finding(pos_sents[0])
```

```python
def NE_flat_list_fn(pos_sents): 
    NE=[]#empty list
    for pos_sent in pos_sents:
        #Single line code here. Call the funtion named_entity_finding(pos_sent) and 
                #append the result to the NE list
        NE.append(named_entity_finding(pos_sent))
    #Single line code here. Flatten the list of lists to the single list NE_flat_list
    NE_flat_list = [item for sublist in NE for item in sublist]
    return NE_flat_list

NE_flat_list_fn(pos_sents)
```

# Task 4 (40 Marks)

Write functions to extract the name of the player, country of origin and date of birth as well as the following relations: team(s) of the player and position(s) of the player.

Hint: Use the re.compile() function to create the extraction patterns

Reference: https://docs.python.org/3/howto/regex.html

```python
def name_of_the_player(doc):
    #code goes here
    name=doc[0:doc.find("(born")] #checks for the person's name
    #The pattern observed is that "(born" is followed by the person's name.
    #return the extracted name.
    return name

def country_of_origin(doc):
    #code goes here
    #country of origin, hence the nationality of the player is in the pattern 
    # that the nationality is followed by the keyword "professional footballer"
    #country=re.findall(r'is [a](.*?) professional footballer',doc)
    #since for the country_of_origin and team_of_the_player same code has to be used
    #I have instead searched for the footballer's nationality.
    country=re.findall(r'is (.*?) professional footballer',doc)
    return country

def date_of_birth(doc):
    #code goes here
    date=doc[doc.index('born')+5:doc.index(')')] #D.O.B is contained between "born" and ")"
    return date

def team_of_the_player(doc):
    #code goes here
    if(len(re.findall(r'the (.*?) national team',doc))>0):#if the list contains more than 1 element
        team=re.findall(r'the (.*?) national team',doc)[0]#extract the first element
    else:
        team=re.findall(r'the (.*?) national team',doc)#else extract the only element.
    return team

def position_of_the_player(doc):
    #code goes here
    position=re.findall(r'midfielder|forward|captain|striker|outfield|winger',doc)
    #The player can be midfielder|forward|captain|striker|outfield|winger or more than 1 of these
    position=list(set(position))#This gives the unique list of the positions of the player
    return position
```

# Task 5 (10 Marks)

Write a function using the outputs from the previous functions to generate JSON-LD output as follows.

Reference: https://json-ld.org/primer/latest/

{ "@id": "http://my-soccer-ontology.com/footballer/name_of_the_player",

    "name": "",
    "born": "",
    "country": "",
    "position": [
        { "@id": "http://my-soccer-ontology.com/position",
            "type": ""
        }
     ]   
     "team": [
        { "@id": "http://my-soccer-ontology.com/team",
            "name": ""
        }   
     ]
}


```python
def generate_jsonld(arg1,arg2,arg3,arg4,arg5):
    #Code goes here
    #Hint: arg1,arg2,..... are the arguments you will be passing to the function
    doc={#key value pairs
        "@id":"http://my-soccer-ontology.com/footballer/",
        "http://schema.org/name":unidecode.unidecode(arg1),
        "http://schema.org/country":arg2,
        "http://schema.org/born":arg3,
        "http://schema.org/team":{"@id":"http://schema.org/name","@type":arg4},
        "http://schema.org/position":{"@id":"http://schema.org/position","@type":arg5}
    }
    #context tells how to interpret the data in doc
    context={
        "name":"http://schema.org/name",
        "country":"http://schema.org/country",
        "born":"http://schema.org/born",
        "team":{"@id":"http://schema.org/team",
               "@type":"@id"},
        "position":{"@id":"http://schema.org/position",
                   "@type":"@id"}
    }
    #compacting the document as per the context
    compacted=jsonld.compact(doc,context)
    compacted.pop("@context")
    return json.dumps(compacted,indent=2)
```

```python
#Extracting player info. with the help of functions.
arg1=name_of_the_player(list_of_doc[1])#contains name
arg2=country_of_origin(list_of_doc[1])#contains nationality
arg3=date_of_birth(list_of_doc[1])#contains date_of_birth
arg4=team_of_the_player(list_of_doc[1])#contains the national team of the player
arg5=position_of_the_player(list_of_doc[1])#contains the position the player plays in.

print(generate_jsonld(arg1,arg2,arg3,arg4,arg5))#printing the json-ld
```

# Task 6 (10 Marks)
Identify one other relation (besides team and player) and write a function to extract this. Also extend the JSON-LD output accordingly.

```python
def won(doc):
    #code goes here
    #This checks if the player has any awards or titles associated with him
    #and returns a list of accolades won.
    awards=[]#empty list
    nonzeroawards=[]#empty list
    #for sentence in the document (ending with a full stop)
    #extract only those sentences with words 'won','title','titles'
    accolades=[sentence + '.' for sentence in doc.split('.') if 'won' or 'title' or 'titles' in sentence]
    #extract words in those sentences (if any) 
    #containing words following "won" and before words like (award|title|titles)
    for i in range(len(accolades)):
        awards.append(re.findall(r'won (.*?) (award|title|titles)',accolades[i]))
    #eliminating empty lists
    for i in range(len(awards)):
        if(len(awards[i])!=0):
            nonzeroawards.append(awards[i][0][0])#extracting only the words between "won" and the (award|title|titles)
    return nonzeroawards
```
